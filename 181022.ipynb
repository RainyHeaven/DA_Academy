```python
import pydotplus
import graphviz
from sklearn.tree import export_graphviz
from IPython.display import Image
from sklearn.metrics import classification_report as report
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier as DTC
import pandas as pd
```

```python
iris = pd.read_csv("C:/Users/stu/git/DA_Academy/iris.csv")

x = iris.drop('Name',axis = 1)
y = iris['Name']

# train, test 로 분리
x_train,x_test,y_train,y_test = tts(x,y,test_size = 0.2) # 20%는 test 80%는 train 로 분리

# 분류 학습모델
classifier = DTC()
classifier.fit(x_train,y_train) 

# 예측
y_predict = classifier.predict(x_test)
y_predict
```

```python
# 예측비교
print(report(y_test,y_predict))
# recall : 문제가 있는 부분
```

```python
classifier.score(x_train,y_train) # 학습용 데이터셋 정확도
classifier.score(x_test,y_test) # 검증용 데이터셋 정확도 

# 임의 test
x1 = [[6.3,2.7,4.9,1.8]]; y1="Iris-virginica"
y1_predict = classifier.predict(x1)
y1_predict
```

```python
import pandas as pd
titanic = pd.read_csv('C:/Users/stu/git/DA_Academy/titanic.csv')
titanic.head()
```

```python
titanic['gender'] = titanic.gender.map({'female':0, 'male':1})
```

```python
titanic.loc[titanic.age.isnull(),:]
```

```python
titanic.age.fillna(titanic.age.median(), inplace=True)
```

```python
titanic.head()
```

```python
# emparked (탑승항구) : C(Cherbourg), Q(Queenstown), S(Southampton)
# one hot encoding = 변수의 값을 0,1 로 치환하는 방법
embarked_dummies = pd.get_dummies(titanic.embarked, prefix='embarked')
```

```python
embarked_dummies
```

```python
embarked_dummies.drop(embarked_dummies.columns[0], axis=1, inplace=True)
```

```python
titanic = pd.concat([titanic, embarked_dummies], axis=1)
titanic
```

```python
titanic.columns
```

```python
feature_cols = ['pclass','gender', 'age','embarked_Q', 'embarked_S']
titanic[feature_cols]
```

```python
x = titanic[feature_cols]
y = titanic.survived
```

```python
#model = DTC(criterion='gini', max_depth=3)
model = DTC(criterion='entropy', max_depth=3)
model.fit(x,y)
```

```python
pd.DataFrame({'feature':feature_cols, 'importance':model.feature_importances_})
```

```python
dot_data = export_graphviz(model, out_file=None, 
                           feature_names=['pclass','gender','age','embarked_Q','embarked_S'],
                           class_names=['0','1'],
                           filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
