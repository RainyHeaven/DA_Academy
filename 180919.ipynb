-  ### DataFrame
    -  `데이터프레임.shape`: 데이터프레임의 모양(차원)
    -  `데이터프레임.columns`: 데이터프레임의 컬럼명
    -  `데이터프레임.dtypes`: 데이터프레임 컬럼들의 데이터타입

-  ### merge()
    -  두 데이터프레임을 특정 컬럼을 기준으로 연결함
    -  `pandas.merge(기준데이터프레임, 연결할 데이터프레임, on=기준컬럼, how=조인방법)`
    -  조인방법
        -  inner: 기본값, 서로 중복되는 컬럼값만 사용
        -  left: 왼쪽 데이터프레임의 컬럼값만 사용
        -  right: 오른쪽 데이터프레임의 컬럼값만 사용
        -  outer: 양쪽 데이터프레임의 컬럼값 모두 사용
    -  기준컬럼의 컬럼명이 서로 다를시엔 left_on, right_on 으로 컬럼 지정   
        ```python
        pd.merge(emp[['name','deptno']], dept[['deptno','dname']], left_on = 'deptno', right_on='deptno')
        ```
        ```
                name	deptno	dname
        0	King	90.0	Executive
        1	Kochhar	90.0	Executive
        2	De Haan	90.0	Executive
        (생략)
        ```
    -  right_index, left_index: 컬럼명이 아닌 인덱스를 기준으로 join
        ```python
        dept = DataFrame({'dname':
                 ['관리팀', '마케팅팀','구매팀','인사팀','경영지원팀','기술지원팀',

        pd.merge(emp[['name','sal','deptno']], dept, left_on='deptno', right_index=True)
        ```
