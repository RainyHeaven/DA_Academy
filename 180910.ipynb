[문제83] stddev(2,3,1,7) 표준편차를 구하세요. stddev함수를 생성하세요.
```
평균 = 관측값의 합 / 데이터수
편차 = 관측값 - 평균
편차 제곱합 = (편차**2)+(편차**2)
분산 = 편차제곱합/데이터수(자유도)
표준편차 = math.sqrt(분산)
import math
math.sqrt(분산)
```
```python
# 하나의 함수로 작성
def stddev(*nums):
    avg = sum(nums)/len(nums)
    dev = [n-avg for n in nums]
    var = sum([n**2 for n in dev])/(len(nums)-1)
    import math
    return math.sqrt(var)
```
```python
# 각각의 함수로 작성
def avg(*nums):
    return sum(nums)/len(nums)

def dev(*nums):
    return [n-avg(*nums) for n in nums]

def var(*nums):
    return sum([n**2 for n in dev(*nums)])/(len(nums)-1)

def stddev(*nums):
    import math
    return math.sqrt(var(*nums))
```
```python
# 람다
def stddev(*nums):

    return math.sqrt(var(nums))
```

[문제84] stats 모듈에 평균, 분산, 표준편차함수를 사용할수 있는 프로그램을 생성하세요.

```python
# stats.py파일 생성
def mean(*nums):
    return sum(nums)/len(nums)

def dev(*nums):
    return [n-mean(*nums) for n in nums]

def variance(*nums):
    return sum([n**2 for n in dev(*nums)])/(len(nums)-1)

def stddev(*nums):
    import math
    return math.sqrt(variance(*nums))
```

# stack
-  FILO(First In Last Out) : 선입후출
-  LIFO(Last In First Out) : 후입선출
-  push
    -  스택의 마지막 데이터 위치에 입력
-  pop
    -  마지막 데이터 위치에서 데이터를 꺼내는 작업(삭제)   

```
stack 변수에 값을 넣는 push 함수 생성
stack 변수의 값을 삭제하는 pop 함수 생성
```
```python
stack = []
def push(x):
    global stack
    stack.insert(len(stack), x)
    return stack

def pop():
    global stack
    if len(stack) >0:
        y = stack[-1]
        del stack[-1]
        print(y)
    else:
        print('stack is empty!!')
```

# 유클리드 알고리즘
-  주어진 두 수 사이에 존재하는 최대공약수(GCD/Greatest Common Divisor)를 구하는 알고리즘
    1.  두 수 m,n 입력 (m>n)
    2.  n이 0이면 m 출력하고 종료
    3.  m이 n으로 나누어 떨어지면, n 출력하고 종료
    4.  m이 n으로 나누어 떨어지지 않으면, m을 n으로 나눈 나머지를 새롭게 m에 대입하고 m과 n을 바꾸고 3번으로 돌아간다.       
-  재귀
    ```python
    def gcd(a,b):
        if b == 0:
            return a
        else:
            return gcd(b, a%b)
    ```
-  반복문
    ```python
    def gcd(a,b):
        if b>a:
            a,b = b,a
        while b!=0:
            c = a%b
            a = b
            b = c
        return a
    ```
