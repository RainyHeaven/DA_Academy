```python
import urllib.request as req
from bs4 import BeautifulSoup as bs

url = 'http://www.skwyverns.com/Wyverns/Players/picther/picther_list.asp'
html = req.urlopen(url)
soup = bs(html, 'html.parser')

imgs = soup.select('#content > div > div.entrie_list > div > div > div.sum > a > img')

for img in imgs:
    req.urlretrieve(('http://www.skwyverns.com'+img.get('src')), 'c:/data/{}.jpg'.format(img.get('alt')))
```

- 네이버 > pay > 쇼핑 > 가전제품 

```python
from selenium import webdriver as wd
from bs4 import BeautifulSoup as bs

user = "" # user id 
mypass = "" # user pw

driver = wd.PhantomJS("c:/data/phantomjs.exe")
driver.implicitly_wait(2)

url_login = "https://nid.naver.com/nidlogin.login"
driver.get(url_login)

inputid = driver.find_element_by_id("id") # id 입력하는 input요소를 찾는다.
inputid.clear() # 입력박스에 있는 텍스트 지우기
inputid.send_keys(user) # user id값을 input창에 작성

inputpw = driver.find_element_by_id("pw")
inputpw.clear()
inputpw.send_keys(mypass)

loginbn = driver.find_element_by_css_selector("input.btn_global[type = submit]") # 로그인 버튼찾기

loginbn.submit() # id pw 전송

driver.get("https://pay.naver.com/introduction/merchant/list?searchMerchantCategoryCode=50000003&searchTapType=merchant&searchSortType=payOrderCount&searchType=merchantName&searchKeyword=")
html = driver.page_source # 태그를 다 가져온다

soup = bs(html,"html.parser") # 가져온 html 컴파일

notices = soup.find_all("table",class_="tb_list tb_store")
for n in notices :
    print(n.text)

driver.quit()
```

### key 제어
- 다음 > 이미지 > 베놈 > 이미지 저장

```python
import urllib.request as req
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
import time

browser = wd.Chrome("c:/data/chromedriver.exe")
browser.get("https://search.daum.net/search?nil_suggest=btn&w=img&DA=SBC&q=")

elem = browser.find_element_by_id("q")
elem.send_keys("아이언맨")
elem.submit()

for i in range(1,2) :
    browser.find_element_by_tag_name("body").send_keys(Keys.END) # end 키( 더보기 ) 를 자동으로 눌러주도록
    time.sleep(2)

html = browser.page_source
soup = bs(html,"html.parser")

params = []
imglist = soup.findAll("img",class_ = 'thumb_img')

for img in imglist :
    params.append(img['src'])

a = 1
for p in params :
    req.urlretrieve(p,"c:/data/"+str(a)+".jpg")
    a += 1

browser.quit()
```

````python
import urllib.request as req
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
import time

browser = wd.Chrome("c:/data/chromedriver.exe")
browser.get("https://search.daum.net/search?nil_suggest=btn&w=img&DA=SBC&q=")

elem = browser.find_element_by_id("q")
elem.send_keys("아이언맨")
elem.submit()

for i in range(1,2) :
    browser.find_element_by_tag_name("body").send_keys(Keys.END) # end 키( 더보기 ) 를 자동으로 눌러주도록
    time.sleep(2)

html = browser.page_source
soup = bs(html,"html.parser")

params = []
imglist = soup.findAll("img",class_ = 'thumb_img')

for img in imglist :
    params.append(img['src'])

a = 1
for p in params :
    req.urlretrieve(p,"c:/data/"+str(a)+".jpg")
    a += 1

browser.quit()
````
