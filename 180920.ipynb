```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

def f(x,y):
    return list(map(lambda x,y:2*x+y, x,y))
```

## pandas
-  ### map()
    -  시리즈에 포함된 각각의 값에 특정 작업을 반복한다.
    -  `시리즈.map(작업, na_action=na처리여부)`
    -  작업
        -  함수: 각각의 값에 함수를 적용
        -  딕셔너리: 시리즈의 값과 딕셔너리의 키를 기준으로 join, 시리즈의 인덱스와 딕셔너리의 값을 값을 갖는 시리즈를 리턴
        -  시리즈: 시리즈1의 값과 시리즈2의 인덱스를 기준으로 join, 시리즈1의 인덱스에 시리즈2의 값을 갖는 시리즈를 리턴
    -  na처리여부: None, ignore
    ```python
    from pandas import Series, DataFrame
    import pandas as pd
    import numpy as np

    x = Series([1,2,3], index=['one','two','three'])
    y = Series(['하나','둘','셋'], index=[1,2,3])
    x.map(y)
    ```
    ```
    one      하나
    two       둘
    three     셋
    dtype: object
    ```
