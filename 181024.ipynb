```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('data/iris.csv')
```

```python
plt.scatter(iris.SepalLength, iris.SepalWidth, s=40)
plt.title('Sepal')
```

```python
plt.scatter(iris.PetalLength, iris.PetalWidth, s=40)
plt.title('Petal')
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(iris.iloc[:,0:4])
model.labels_
```

```python
colormap=np.array(['red','blue','black'])
plt.scatter(iris.SepalLength, iris.SepalWidth, s=40,
            c=colormap[model.labels_])
plt.title('K Mean Classification')
```

```python
colormap=np.array(['red','blue','black'])
plt.scatter(iris.PetalLength, iris.PetalWidth, s=40,
            c=colormap[model.labels_])
plt.title('K Mean Classification')
```

```python
academy = pd.read_csv('data/academy.csv', encoding='cp949', index_col='')
```

```python
academy
```

```python
km = KMeans(n_clusters=4)
km.fit(academy.iloc[:,1:5])
colormap=np.array(['red','blue','black', 'purple'])
plt.scatter(academy., academy., s=40,
            c=colormap[km.labels_])
plt.title('K Mean Classification')
```

```python
colormap=np.array(['red','blue','black', 'purple'])
plt.scatter(academy.iloc[:,2], academy.iloc[:,3],c=colormap[km.labels_], s=30)
centers = pd.DataFrame(km.cluster_centers_)
plt.scatter(centers.iloc[:,0], centers.iloc[:,1], s=50, marker='D', c='g')
plt.show()
```

```python
# k   (inertia) 
inertia = []
ks = range(1,10)
for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(academy.iloc[:,2:4])
    inertia.append(model.inertia_)
```

```python
#     
# inertia value :       
inertia
```

```python
plt.plot(ks, inertia, '-o')
plt.xlabel('number of cluster k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
