
[문제28] 숫자를 입력값으로 받습니다. 
 만약 숫자가 이외의 값이 들어 오면 "숫자 이외의 문자가 포함되어 있습니다." 아니면 숫자 출력하세요.
```
char = input()
if char.isnumeric():
    print(char)
else:
    print('숫자 이외의 문자가 포함되어 있습니다')
```

[문제29] 1부터 100까지 합을 구하세요.
```
i = 1
sum = 0
while i <= 100:
    sum += i
    i += 1
print(sum)
```

[문제30] 단을 입력값으로 받아서 구구단을 출력하세요.
```
# for문
dan = int(input())
print('구구단 '+str(dan)+'단 입니다.')
for i in range(1,10):
    print('{} x {} = {}'.format(dan, i, dan*i))

# while문
dan = int(input())
print('구구단 '+str(dan)+'단 입니다.')
i = 1
while i <= 9:
    print('{} x {} = {}'.format(dan, i, dan*i))
    i += 1
```

### 중첩리스트
```python
x = []
type(x)
```
> list

```python
x = [1,2,3,['a','b','c'],4,5]
x[2:5]
```
> [3, ['a', 'b', 'c'], 4]

```python
x[3]
```
> ['a', 'b', 'c']

```python
x[3][0]
```
> 'a'

```python
x[3][0:2]
```
> ['a', 'b']

```python
x[3][-1]
```
> 'c'

```python
x[3].append('d')
x
```
> [1, 2, 3, ['a', 'b', 'c', 'd'], 4, 5]

```python
del x[3][3]
x
```
> [1, 2, 3, ['a', 'b', 'c'], 4, 5]

```python
x[3][0] = x[3][0].upper()
x
```
> [1, 2, 3, ['A', 'b', 'c'], 4, 5]

```python
x[0] = x[0]*10
```
> [10, 2, 3, ['A', 'b', 'c'], 4, 5]

```python
x[3][0] = x[3][0]*2
```
> [10, 2, 3, ['AA', 'b', 'c'], 4, 5]

### clear()
-  변수 내의 값을 모두 삭제
-  변수.clear()
```python
x.clear()
```
> []

### sort()
-  변수 내의 값을 기준으로 정렬.
-  기본값은 오름차순
-  변수.sort()
```python
x = [1,5,3,4,2]
x.sort() # 오름차순
x
```
> [1, 2, 3, 4, 5]
-  변수.sort(reverse=True)
```python
x.sort(reverse=True) # 내림차순
x
```
> [5, 4, 3, 2, 1]

### reverse()
-  리스트 인덱스 순서를 반대로 뒤집을 때 사용
-  변수.reverse()
```python
x = [1,5,3,4,2]
x.reverse()
x
```
> [2, 4, 3, 5, 1]

## tuple
-  리스트자료형과 똑같지만 수정, 삭제, 추가 불가능
-  리스트 [ ] / 튜플 ( )
```python
list1 = []
tuple1 = ()
tuple2 = 10,20
tuple3 = (1,)
tuple4 = (1,2,3,4)
tuple5 = (1,2,3,('a','b'))
```
-  튜플 변수 값은 수정 불가   
```python
tuple5[0] = 10
```
> TypeError: 'tuple' object does not support item assignment

-  튜플 변수 값은 삭제 불가   
```python
del tuple5[0]
```
> TypeError: 'tuple' object doesn't support item deletion

-  튜플 변수 값은 추가 불가   
```python
tuple5.append(4)
```
> AttributeError: 'tuple' object has no attribute 'append'

-  튜플을 합치려면 새로운 변수로 생성해야 함   
```python
x = (1,2,3)
y = (4,5,6)
```
```python
z = x + y
x = x+(4,)
z
x
```
> (1, 2, 3, 4, 5, 6)   
> (1, 2, 3, 4)

-  새로 생성된 튜플은 이전 튜플과는 다른 객체
```python
x = (1,2,3)
id(x)
```
> 81557400

    ```python
    x = x+(4,)
    id(x)
    ```
    > 81491112

-  튜플 변수 값은 정렬 불가   
```python
x = (2,3,4,1)
x.sort(x)
```
> AttributeError: 'tuple' object has no attribute 'sort'

-  정렬된 값을 다른 변수에 저장    
```python
z = sorted(x)
z
```
> [1, 2, 3, 4]

## dictionary
-  key를 통해 value를 확인하는 자료형

| key | value |
| --- | --- |
| 이름 | 홍길동 |
| 전화번호 | 01012345678 |
| 주소 | 경기도 |

```python
dic = {'이름':'홍길동', '전화번호':'01012345678', '주소':'경기도'}
type(dic)
```
> dict

```python
dic['이름']
```
> '홍길동'

```python
sports = {'축구':'메시', '농구':'커리','야구':'박찬호'}
sports['축구']
```
> '메시'

```python
sports['컬링'] = '김영미'
sports
```
> {'축구': '메시', '농구': '커리', '야구': '박찬호', '컬링': '김영미'}

### keys()
-  딕셔너리의 모든 키 값들을 리턴
-  딕셔너리.keys()
```python
sports['컬링'] = ['김은정','김경애','김영미','김선영','김초희']
sports.keys()
```
> dict_keys(['축구', '농구', '야구', '컬링'])

### values()
-  딕셔너리의 모든 키에 할당된 값들을 리턴
-  딕셔너리.values()
```python
sports.values()
```
> dict_values(['메시', '커리', '박찬호', ['김은정', '김경애', '김영미', '김선영', '김초희']])

### items()
-  딕셔너리의 모든 키와 키에 할당된 값들을 묶어서 리턴
-  딕셔너리.items()
```python
sports.items()
```
> dict_items([('축구', '메시'), ('농구', '커리'), ('야구', '박찬호'), ('컬링', ['김은정', '김경애', '김영미', '김선영', '김초희'])])

### get()
-  딕셔너리에서 해당 키에 할당된 값들을 리턴
-  딕셔너리.get(키)
```python
sports['농구']
```
> '커리'

    ```python
    sports.get('농구')
    ```
   > '커리'

    ```python
    sports['봅슬레이'] # 없는키로 찾을 시 오류 발생
    ```
   > KeyError: '봅슬레이'

    ```python
    sports.get('봅슬레이') #없는 키로 찾을 시 none 리턴
    ```
   > 

-  딕셔너리의 요소 확인
```python
'컬링' in sports.keys()
```
> True

    ```python
    '봅슬레이' in sports.keys()
    ```
> False

    ```python
    '메시' in sports.values()
    ```
> True

    ```python
    '김영미' in sports.values()
    ```
> False

    ```python
    ['김은정','김경애','김영미','김선영','김초희'] in sports.values()
    ```
> True

    ```python
    del sports['야구']
    sports
    ```
> {'축구': '메시', '농구': '커리', '컬링': ['김은정', '김경애', '김영미', '김선영', '김초희']}

    ```python
    sports['축구'] = []
    sports
    ```
> {'축구': [], '농구': '커리', '컬링': ['김은정', '김경애', '김영미', '김선영', '김초희']}

    ```python
    sports.clear()
    sports
    ```
> {}

    ```python
    x = sports.values()
    type(x)
    ```
> dict_values

    ```python
    list(x)
    ```
> []

# set(집합)
-  list type과 유사
-  index가 없다
-  중복을 허용하지 않는다
    ```python
    x = {1,1,1,2,3,2,3}
    x
    ```
> {1, 2, 3}

    ```python
    type(x)
    ```
> set

# 합집합
```python
x = {1,1,1,2,3,2,3}
y = {2,3,4,5}
```

-  ### union()
    -  변수의 집합과 지정한 집합의 합집합 리턴
    -  변수.union(집합)
    ```python
    x.union(y)
    ```
> {1, 2, 3, 4, 5}

-  x|y
    -  두 집합의 합집합 리턴
    -  집합1|집합2
    ```python
    x|y
    ```
> {1, 2, 3, 4, 5}


# 교집합

-  ### intersection()
    -  변수의 집합과 지정한 집합의 교집합 리턴
    -  변수.intersection(집합)
    ```python
    x.intersection(y)
    ```
> {2, 3}

-  x&y
    -  두 집합의 교집합 리턴
    -  집합1&집합2
    ```python
    x&y
    ```
    > {2, 3}

# 차집합

-  ### difference()
    -  변수의 집합과 지정한 집합의 차집합 리턴
    -  변수.difference(집합)
    ```python
    x.difference(y)
    ```
    > {1}
    
    ```python
    y.difference(x)
    ```
    > {4, 5}

-  x-y
    -  두 변수의 차집합 리턴
    -  x - y
    ```python
    x-y
    ```
    > {1}
    
    ```python
    y-x
    ```
    > {4, 5}

-  집합의 인덱싱
    ```python
    1 in x
    ```
> True

    ```python
    6 in x
    ```
> False

    -  set에는 index가 없음
    ```python
    x[0]
    ```
> TypeError: 'set' object does not support indexing

### remove()
-  집합 변수의 값을 삭제
-  변수.remove(값)
```python
x.remove(1)
x
```
> {2, 3}


### add()
-  집합 변수에 값을 입력
-  변수.add(값)
    ```python
    x.add(1)
    ```
> {1, 2, 3}

### update()
-  집합 변수에 여러개의 값을 추가
-  변수.update(값)
    ```python
    x.update([5,6,7])
    x
    ```
> {1, 2, 3, 5, 6, 7}

-  데이터 타입
    ```python
    a = []
    b = ()
    c = {}
    s = set()
    ```
    ```python
    type(a)
    ```
> list

    ```python
    type(b)
    ```
> tuple

    ```python
    type(c)
    ```
> dict

    ```python
    type(s)
    ```
> set

# bool
-  참(True), 거짓(False)를 나타내는 자료형

```python
x = True
type(x)
```
> bool

```python
y = False
type(y)
``` 
> bool

```python
x == y
```
> False

```python
1 == 1
```
> True

```python
2 > 1
```
> True

```python
1 != 2
```
> True

```python
not 1
```
> False

```python
not -2
```
> False

```python
not 0
```
> True

```python
not None
```
> True

# True 표현방법

```python
bool(1)
bool(-1)
bool('python')
bool([1,2,3])
bool((1,2,3))
bool({1,2})
```

# False 표현방법

```python
bool(0)
bool([])
bool(())
bool({})
bool(None)
bool('')
```

# if문
-  if 조건문:   
    수행해야 할 문장

```python
if 1:
    print('참')
    print('오늘 하루도 행복하자')
```
> 참   
오늘 하루도 행복하자

```python
if 0:
    print('참')
    print('오늘 하루도 행복하자')
else:
    print('거짓')
    print('그냥 사는거지 뭐...')
```
> 거짓   
그냥 사는거지 뭐...

```python
if 0:
    print('참')
    print('오늘 하루도 행복하자')
else:
    print('거짓')
    print('그냥 사는거지 뭐...')
```
> 거짓   
그냥 사는거지 뭐...

```python
x = []
if x:
    print('참')
    print('오늘 하루도 행복하자')
else:
    print('거짓')
    print('그냥 사는거지 뭐...')
```
> 거짓
그냥 사는거지 뭐...

# and: 조건을 순서대로 평가
```python
x = 0
if x > 10 and 1 / x :
    print('x는 10보다 크다')
else:
    print('x는 10보다 작다')
```
> x는 10보다 작다

```python
x = 0
if 1 / x and x > 10:
    print('x는 10보다 크다')
else:
    print('x는 10보다 작다')
```
> ZeroDivisionError: division by zero

# &: 조건식을 모두 평가
```python
x = 0
if 1 >10 & 1 / x:
    print('x는 10보다 크다')
else:
    print('x는 10보다 작다')
```
> ZeroDivisionError: division by zero

# or : 조건을 순서대로 평가
```python
x = 0
if x < 10 or 1 / x:
    print('x는 10보다 크다')
else:
    print('x는 10보다 작다')
```
> x는 10보다 크다

# | : 조건을 모두 평가
```python
x = 0
if x < 10 | 1 / x:
    print('x는 10보다 크다')
else:
    print('x는 10보다 작다')
```
> ZeroDivisionError: division by zero

```python
score = 88
if 90 <= score <= 100:
    print('A')
elif 80 <= score < 90:
    print('B')
elif 70 <= score < 80:
    print('C')
elif 60 <= score < 70:
    print('D')
else:
    print('F')
```
> B

# 반복문
### while문
-  조건이 참인동안에 while문을 반복해서 수행한다
-  while 조건문:    
    반복수행해야 할 문장
