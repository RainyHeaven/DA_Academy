# 브로드캐스팅(broadcasting)
-  서로 다른 차원을 가진 행렬끼리의 연산이 가능하도록 작은 행렬을 큰 행렬의 차원에 맞도록 반복하여 행렬의 차원을 일치시키는 방법
-  작은 차원의 n배가 큰 차원과 일치할 때만 작동함
    ```python
    # 수동 브로드캐스팅
    obj1 = np.arange(15).reshape(5,3)
    obj2 = np.arange(3)
    obj1 + obj2.repeat(5).reshape(3,5).T
    ```
    ```
    array([[ 0,  2,  4],
       [ 3,  5,  7],
       [ 6,  8, 10],
       [ 9, 11, 13],
       [12, 14, 16]])
    ```

# 정렬
### sort_index()
-  index 기준 오름차순 정렬
-  객체.sort_index(ascending=오름차순여부, axis=축)
-  축: 0(row), 1(column)
    ```python
    df = DataFrame(np.arange(8).reshape(2, 4), 
                   index=['two', 'one'], 
                   columns=['d','a','c','b'])
    
    df.sort_index(ascending=False, axis=1)
    ```
    ```
    	d	c	b	a
    two	0	2	3	1
    one	4	6	7	5
    ```
    
### sort_values()
-  value 기준 오름차순 정렬
-  객체.sort_values(ascending=오름차순여부, by = 기준, axis=축)
-  축: 0(row), 1(column)
    ```python
    df.sort_values(by = 'b', axis=0, ascending=False)
    ```
    ```
        d	a	c	b
    one	4	5	6	7
    two	0	1	2	3
    ```
