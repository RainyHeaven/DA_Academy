## 문자함수
### startswith()
-  해당 변수가 지정한 문자열로 시작되는지 확인
-  변수명.startswith(문자열)

### endswith()
-  해당 변수가 지정한 문자열로 끝나는지 판단
-  변수명.endswith(문자열)

### find()
-  해당 변수에 지정한 문자열이 존재하는 위치 리턴. 존재하지 않을 시 -1 리턴
-  변수명.find(문자열)

### count()
-  해당 변수에 지정한 문자열이 등장하는 횟수 리턴
-  변수명.find(문자열)

### upper()
-  해당 변수의 문자열을 대문자로 변환
-  변수명.upper()

### lower()
-  해당 변수의 문자열을 소문자로 변환
-  변수명.lower()

### capitalize()
-  해당 변수의 문자열에서 첫글자는 대문자, 나머지는 소문자로 변환
-  변수명.capitalize()

### title()
-  해당 변수의 문자열에서 매 단어의 첫글자를 대문자로, 나머지를 소문자로 변환
-  변수명.title()

### swapcase()
-  해당 변수의 문자열에서 대문자는 소문자로, 소문자는 대문자로 변환
-  변수명.swapcase()

### center()
-  지정한 길이의 빈 문자열을 만들고 해당 변수의 문자열을 정 가운데에 삽입
-  변수명.center(길이)

### ljust()
-  지정한 길이의 빈 문자열을 만들고 해당 변수의 문자열을 왼쪽 끝에 삽입
-  변수명.ljust(길이)

### rjust()
-  지정한 길이의 빈 문자열을 만들고 해당 변수의 문자열을 오른쪽 끝에 삽입
-  변수명.rjust()

### strip()
-  해당 변수의 문자열에서 양 끝의 공백을 제거
-  변수명.strip()

### lstrip()
-  해당 변수의 문자열에서 왼쪽 끝의 공백을 제거
-  변수명.lstrip()

### rstrip()
-  해당 변수의 문자열에서 오른쪽 끝의 공백을 제거
-  변수명.rstrip()

### isalpha()
-  해당 변수의 문자열이 알파벳으로 이루어져있는지 확인
-  변수명.isalpha()

### isalnum
-  해당 변수의 문자열이 알파벳, 숫자로 이루어져있는지 확인
-  변수명.isalnum()

### isnumeric()
-  해당 변수의 문자열이 숫자로 이루어져있는지 확인
-  변수명.isnumeric()

### replace()
-  해당 변수의 문자열에서 지정한 문자열을 찾아 새로운 문자열로 변경
-  변수명.replace(찾을문자열, 바꿀문자열)

### index()
-  해당 변수의 문자열에서 지정한 문자열이 존재하는 위치를 앞에서부터 찾아서 리턴. 존재하지 않을 시 오류
-  변수명.index(문자열)

### split()
-  해당 변수의 문자열을 지정한 문자열을 기준으로 분할, 리스트 형태로 리턴
-  변수명.split(문자열)

### join()
-  특정 문자열을 해당 변수의 문자열 사이에 추가
-  넣을문자열.join(변수)
